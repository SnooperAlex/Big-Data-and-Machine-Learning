```python
import numpy
import matplotlib
from sklearn.linear_model import LinearRegression
plt = matplotlib.pyplot
```

```python
x_data = numpy.load('./x_values.npy')
y_data = numpy.load('./y_values.npy')
x_testData = numpy.load('./test_x_values.npy')

lreg = LinearRegression().fit(x_data, y_data)
y_pred = lreg.predict(x_testData)

plt.scatter(x_data, y_data)
plt.plot(x_testData, y_pred)
plt.show

y_array = numpy.array([0.48])
y_pred = lreg.predict(y_array.reshape(1, -1))
print(y_pred)
```

Task 3.2

```python
from sklearn.decomposition import PCA
import random
from sklearn.model_selection import train_test_split
```

```python
wine_data = numpy.load('./wineData.npy')
wine_label = numpy.load('./wineLabels.npy')


train_data, test_data, train_labels, test_labels = train_test_split(wine_data, wine_label, train_size = 0.8)
```

```python
plt.scatter(train_data[:,0], train_data[:,1], 25,train_labels)
plt.scatter(test_data[:,0], test_data[:,1], 25, test_labels, marker = "x")
plt.show
```

```python
pca = PCA(n_components = 2)
pca.fit(train_data)
Xt = pca.transform(train_data)
Xts = pca.transform(test_data)

plt.scatter(Xt[:,0], Xt[:,1], 25,train_labels)
plt.scatter(Xts[:,0], Xts[:,1], 25, test_labels, marker = "x")
plt.show
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(train_data, train_labels)
lda_x = lda.transform(train_data)
lda_y = lda.transform(test_data)
```

```python
plt.scatter(lda_x[:,0], lda_x[:,1], 25,train_labels)
plt.scatter(lda_y[:,0], lda_y[:,1], 25, test_labels, marker = "x")
plt.show
```

Task 3.4

```python
mean_centre = numpy.mean(train_data)
train_data_centred = train_data
for i in range(len(train_data)):
  train_data_centred[i] = train_data[i] - mean_centre

mean_centre_test = numpy.mean(test_data)
test_data_centred = test_data
for i in range(len(test_data)):
  test_data_centred[i] = test_data[i] - mean_centre_test
```

```python
u, s, vh = numpy.linalg.svd(train_data_centred)
u2, s2, vh2 = numpy.linalg.svd(test_data_centred)
```

```python
projection_matrix = vh[1:]
projected_train_data = train_data_centred @ numpy.transpose(projection_matrix)

projection_matrix2 = vh2[1:]
projected_test_data = test_data_centred @ numpy.transpose(projection_matrix2)
```

```python
plt.scatter(projected_train_data[:,0], projected_train_data[:,1], 25,train_labels)
plt.scatter(projected_test_data[:,0], projected_test_data[:,1], 25, test_labels, marker = "x")
plt.show
```
