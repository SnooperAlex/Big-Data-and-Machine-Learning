```python
import numpy
import matplotlib
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
data = numpy.load('./Iris_data.npy')
labels = numpy.load('./Iris_labels.npy')

print(labels)
```

```python
l = []
for i in range(len(data)):
  l.append(data[i][0])

k = []
for i in range(len(data)):
  k.append(data[i][1])


plt = matplotlib.pyplot
matplotlib.pyplot.scatter(l,k,50,labels)
plt.title("Fisher Iris Sepal")
plt.xlabel("Sepal height")
plt.ylabel("Sepal width")
plt.show()
```

K-Means


Following code will scatter a plot using k-means method of clustering and grouping them correctly

```python
kMeans = KMeans(n_clusters=3)
kModel = kMeans.fit_predict(data)
kMeans.labels_
```

```python
kMeans.cluster_centers_
```

```python
(kMeans.cluster_centers_[: 1]).shape
```

```python
plt.scatter(data[kModel == 0, 0],data[kModel == 0, 1],label = "iris s")
plt.scatter(data[kModel ==1, 0],data[kModel ==1, 1],label = "iris v")
plt.scatter(data[kModel ==2, 0],data[kModel ==2, 1],label = "iris vi")
plt.scatter(kMeans.cluster_centers_[:, 0],kMeans.cluster_centers_[:, 1],marker = "x", label = "centroids")
plt.title("Iris table")
plt.legend()
```

Gaussian Mixture




Following code will scatter a plot using gaussian mixture model of clustering and grouping them accurately


```python
gm = GaussianMixture(n_components=3, random_state=0).fit(data)
gm.means_
```

```python
g_label = gm.predict(data)

plt.scatter(data[:, 0], data[:, 1], c = g_label, s=40, cmap='viridis');
```

```python
g_label_proba = gm.predict_proba(data)


plt.scatter(data[:, 0], data[:, 1], c = g_label_proba[:, 0], s=50);
```

```python
g_label_proba = gm.predict_proba(data)


plt.scatter(data[:, 0], data[:, 1], c = g_label_proba[:, 1], s=50);
```

```python
g_label_proba = gm.predict_proba(data)


plt.scatter(data[:, 0], data[:, 1], c = g_label_proba[:, 2], s=50);
```

K-Means

```python
kMeans = KMeans(n_clusters=4)
kModel = kMeans.fit_predict(datax)
kMeans.labels_
kMeans.cluster_centers_
(kMeans.cluster_centers_[: 1]).shape

plt.scatter(datax[kModel == 0, 0],datax[kModel == 0, 1],label = "iris s")
plt.scatter(datax[kModel ==1, 0],datax[kModel ==1, 1],label = "iris v")
plt.scatter(datax[kModel ==2, 0],datax[kModel ==2, 1],label = "iris vi")
plt.scatter(datax[kModel ==3, 0],datax[kModel ==3, 1],label = "iris vi")
plt.scatter(kMeans.cluster_centers_[:, 0],kMeans.cluster_centers_[:, 1],marker = "x", label = "centroids")
plt.title("your k means title")
plt.legend()
```
